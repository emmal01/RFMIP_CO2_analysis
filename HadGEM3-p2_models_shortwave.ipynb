```python
## import packages and intake catalog

#import packages
import numpy as np
import xarray as xr
import netCDF4 as nc
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import seaborn as sns
import pandas as pd
import colorcet as cc
import holoviews as hv
from colorcet.plotting import swatch, swatches, candy_buttons
hv.extension('bokeh')
import warnings
warnings.simplefilter('ignore',  xr.SerializationWarning)

#copy your local directory here
os.chdir('/Users/emmalayton/Desktop/senior_thesis/RFMIP-IRF_intake_catalogue/')

#open the catalog and experiment info
from intake import open_catalog
cat = open_catalog(os.getcwd() + '/main.yml')
extrainfo = xr.open_dataset(os.getcwd() + '/multiple_input4MIPs_radiation_RFMIP_UColorado-RFMIP-1-2_none.nc').to_dataframe()
```

```python
## probe un-avgd TOA rsd values by model and experiment
#make a dataset of the pre-averaged benchmark model values
benchmark_data = {}
for m in list(cat.benchmark):
    for r in list(cat.benchmark[m]):
        components = list(cat.benchmark[m][r])
        benchmark_data[f'{m}_{r}'] = xr.concat([(cat.benchmark[m][r][f].to_dask()) 
                                        for f in components], 
                                        dim="field", data_vars = "different")
        if "rsd" in components:
            benchmark_data[f'{m}_{r}'] = benchmark_data[f'{m}_{r}'].assign(rsn=benchmark_data[f'{m}_{r}'].data_vars['rsd']-benchmark_data[f'{m}_{r}'].data_vars['rsu'])

benchmark_data = {k: benchmark_data[k].assign(rln=benchmark_data[k].data_vars['rld']-benchmark_data[k].data_vars['rlu']) for k in benchmark_data.keys()}
benchmark_data = {k: benchmark_data[k].assign_coords(expt=extrainfo['expt_label'].unique()) for k in benchmark_data.keys()}

```

```python
#plots comparing the HadGEM3p2f2 and HadGEM3p2f3 rsd values at top of atmosphere to LBLRTM
fig, ax = plt.subplots(1,2, figsize=(16,8))
ax[0].hist(benchmark_data["LBLRTM_p1f1"].rsd.sel(level=0).values.flatten())
ax[0].hist(benchmark_data["HadGEM3_p2f2"].rsd.sel(level=0).values.flatten())
ax[0].set_title("LBLRTM_p1f1 vs HadGEM3_p2f2 top of atmosphere rsd values")
ax[0].legend(["LBLRTM_p1f1","HadGEM3_p2f2"])

ax[1].hist(benchmark_data["LBLRTM_p1f1"].rsd.sel(level=0).values.flatten())
ax[1].hist(benchmark_data["HadGEM3_p2f3"].rsd.sel(level=0).values.flatten())
ax[1].set_title("LBLRTM_p1f1 vs HadGEM3_p2f3 top of atmosphere rsd values")
ax[1].legend(["LBLRTM_p1f1","HadGEM3_p2f3"])
```
